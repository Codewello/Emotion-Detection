```python
from nrclex import NRCLex
import plotly.express as px
import pandas as pd
import nltk
nltk.download('punkt')
```

```python
#import the data set using panda
df = pd.read_csv('/content/training.csv')


#display first 5 rows
df.head()
```

```python
#create new data frame using NRCLex

EmotionDataframe = NRCLex(' '.join(df['text']))
```

```python
#make dict for emotions sort it by the value and use it as the index
EmotionDataframe = pd.DataFrame.from_dict(EmotionDataframe.affect_frequencies, orient='index').sort_values(by=0, ascending=False).reset_index()

#the columens are
EmotionDataframe.columns = ['Emotion', 'Frequency']

#display it
EmotionDataframe
```

```python
#create pi figure using px plotly

fig = px.pie(EmotionDataframe,
             values = 'Frequency', names='Emotion',
             title='Emotion Frequency For Reviews',
             hover_data=['Emotion'], labels={'Emotions':'Emotion'})


fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```
